```python
from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

text = "Natural language processing is fascinating."
inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
print(outputs.last_hidden_state)
```

```python
outputs = model(**inputs)           # BERT forward 실행 (**inputs : 키워드 인자를 묶어서 전달)
print(outputs.last_hidden_state)    # 각 토큰의 마지막 레이어 은닉 벡터 출력
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# 사전학습 GPT-2 토크나이저 , 모델 로드
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

input_text = "Artificial Intelligence is"
inputs = tokenizer(input_text, return_tensors="pt")
outputs = model.generate(inputs['input_ids'], max_length=20, num_return_sequences=1)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
outputs = model.generate(
    inputs['input_ids'],         # 압력 토큰 ID
    max_length = 20,            # (프롬프트 포함) 총 토큰 길이 제한
    num_return_sequences = 1    # 생성 문장 갯수 
)
```

```python
# 생성된 id -> text로 복원 후 출력(디코딩시 특수토큰은 출력 문자열에서 제거)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
