```python
import torch
import torch.nn as nn 
```

```python
# мқҙм „ stepмқҳ context(м–ҙн…җм…ҳ кІ°кіј)лҘј лӢӨмқҢ step мһ…л Ҙм—җ л¶ҷм—¬ LSTMм—җ л„ЈлҠ” Input Feeding л””мҪ”лҚ”
class InputFeedingDecoder(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):  
        super(InputFeedingDecoder, self).__init__()
        self.hidden_size = hidden_size                              # л””мҪ”лҚ” мқҖлӢү м°Ёмӣҗ м ҖмһҘ
        self.embedding = nn.Embedding(output_size, input_size)      # м¶ңл Ҙ нҶ нҒ° ID -> мһ…л Ҙ мһ„лІ л”©(input_size)
        self.lstm = nn.LSTM(input_size + hidden_size, hidden_size)  # (мһ„лІ л”© + context) -> hidden_size LSTM
        self.fc = nn.Linear(hidden_size, output_size)               # hidden -> м–ҙнңҳ нҒ¬кё°(output_size) лЎңм§“
        
    # мһ…л Ҙ нҶ нҒ°/мқҙм „ hidden/мқҙм „ contextлҘј л°ӣм•„ лӢӨмқҢ нҶ нҒ° лЎңм§“кіј hiddenмқ„ л°ҳнҷҳ
    def forward(self, input, hidden, context):
       embedded = self.embedding(input).unsqueeze(0)                    # (B, ) -> (1, B, input_size)
       lstm_input = torch.cat((embedded, context.unsqueeze(0)), dim=2)  # мһ„лІ л”©кіј contextлҘј feature м°ЁмӣҗмңјлЎң кІ°н•©
       output, hidden = self.lstm(lstm_input, hidden)                   # LSTM 1-step мҲҳн–ү -> output(1, B, H), hidden((1, B, H),(1, B, H))
       output = self.fc(output.squeeze(0))                              # (1, B, H) -> (B, H) -> (B, output_size)лЎң ліҖнҷҳ
       return output, hidden                                            # лЎңм§“кіј лӢӨмқҢ hiddenк°’ л°ҳнҷҳ
        
```

```python
# Input Feeding Decoder лҚ”лҜё мһ…л ҘмңјлЎң 1-step мӢӨн–ү нҷ•мқё
decoder = InputFeedingDecoder(input_size = 10, hidden_size = 20, output_size = 30)  # л””мҪ”лҚ” мғқм„ұ(мһ„лІ л”©: 10, hidden: 20, vocab:30)
hidden = (torch.zeros(1, 1, 20), torch.zeros(1, 1, 20))     # мҙҲкё° hidden state(h, c) : (num_layers, B, H)
context = torch.zeros(1, 20)                                # мқҙм „ context лІЎн„° (B, H)
input_token = torch.tensor([5])                             # мһ…л Ҙ нҶ нҒ° ID(B, )

output, hidden = decoder(input_token, hidden, context)      # л””мҪ”лҚ” 1-step мӢӨн–ү
output.shape, hidden[0].shape, hidden[1].shape              г…Ғ # output(B, vocab_size), hidden(h), hidden(c) : (num_layers, B, H)
```
